```python
import numpy as np
import seaborn as sns
from scipy.stats import linregress

%matplotlib inline
```

```python
m = 10000

np.random.seed(0)
data = np.random.uniform(size=(m, 2))
```

```python
sns.regplot(data[:, 0], data[:, 1])
```

```python
dist = np.abs(data[:, 0] - data[:, 1])
ind = np.argsort(dist)
data = data[ind]
```

```python
res = []

for i in range(2, m):
    res.append(linregress(data[:i, 0], data[:i, 1]).pvalue)
    
res = np.array(res)
```

```python
sns.regplot(np.arange(2, m), res)
```

```python
i = np.where(res < 0.05)[0].max() + 2
linregress(data[:i, 0], data[:i, 1]).pvalue
```
